```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fun(x):
    
    y = 3*x**2
    
    return y
```

```python
x = np.linspace(0, 1, 1000)
y = fun(x)

plt.plot(x, y, '-')
plt.xlim([0, 5])
```

```python
# define limits

xlim = [0, 1]
ylim = [0, 4]
n_steps = 1000000
x_new_above = []
y_new_above = []

x_new_below = []
y_new_below = []

for i in range(n_steps):
    
    xtest = np.random.uniform(0, 1)
    ytest = np.random.uniform(0, 3)
    
    y_val = fun(xtest)
    
    if y_val>=ytest:
        
        x_new_above.append(xtest)
        y_new_above.append(ytest)
    
    else:
        
        x_new_below.append(xtest)
        y_new_below.append(ytest)


x_new_above = np.array(x_new_above)  
y_new_above = np.array(y_new_above)  

x_new_below = np.array(x_new_below)  
y_new_below = np.array(y_new_below) 


```

```python
plt.plot(x_new_below, y_new_below, 'ro')
plt.plot(x_new_above, y_new_above, 'bo' )
plt.plot(x, y, 'k-', linewidth = 2)

percent_above = len(x_new_below)/n_steps
percent_below = len(x_new_above)/n_steps

area_under_curve = 3*percent_below
print(area_under_curve)
```
